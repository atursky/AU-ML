```python
import os
import polars as pl

folder = r"C:\Users\ANTRS\Downloads\00_DS"
accidents_file = "road.csv"  # Ensure the file name is correct

# Construct the full file path
file_path = os.path.join(folder, accidents_file)

# Check if the file exists
if os.path.exists(file_path):
    # Read the CSV file into a polars DataFrame
    df = pl.read_csv(file_path)
    # Display the first few rows of the DataFrame
    print(df.head())
else:
    print(f"File not found: {file_path}")
```

```python
import pandas as pd

# Load the key.csv file
key_file_path = "C:/Users/ANTRS/Downloads/00_DS/dataset/key.csv"
key_df = pd.read_csv(key_file_path)

# Convert the key.csv to a mapping dictionary
mapping_dict = {}
for attribute in key_df['Attribute'].unique():
    attribute_mapping = key_df[key_df['Attribute'] == attribute].set_index('Code')['Label'].to_dict()
    mapping_dict[attribute] = attribute_mapping
```

```python
def apply_mapping(df, mapping_dict):
    for col, mapping in mapping_dict.items():
        if col in df.columns:
            df = df.with_columns(pl.col(col).replace(mapping).alias(col))
        else:
            # print(f"Column '{col}' not found in DataFrame")
            pass
    return df

df = apply_mapping(df, mapping_dict)
```

```python
import polars as pl
import hvplot.polars
import panel as pn
import holoviews as hv

hv.extension('bokeh')

# Assuming accidents_df is your Polars DataFrame and is already loaded

# Group the data by year and accident severity, and count occurrences
severity_counts = (
    accidents_df.group_by(['accident_year', 'accident_severity'])
    .agg(pl.count('accident_index').alias('count'))  # Ensure we have the count column
    .sort(['accident_year', 'accident_severity'])
)

# Define a color palette for severity levels
color_palette = {1: '#FFA07A', 2: '#FFD700', 3: '#98FB98'}  # Example colors: Light Salmon, Gold, Pale Green

# Function to create the chart based on selected severities
@pn.depends(severity_toggle.param.value)
def filtered_chart(severity_levels):
    # Check if all options are deselected
    if not severity_levels:
        severity_toggle.value = [1, 2, 3]  # Reset to initial state
        severity_levels = [1, 2, 3]
    
    filtered_data = severity_counts.filter(pl.col('accident_severity').is_in(severity_levels))
    return filtered_data.hvplot.bar(
        x='accident_year',
        y='count',
        by='accident_severity',
        stacked=True,
        title='Accident Severity Counts by Year',
        xlabel='Year',
        ylabel='Number of Accidents',
        height=500,
        width=800,
        color=[color_palette[sev] for sev in severity_levels]
    ).opts(
        hv.opts.Bars(
            line_color='white', 
            line_width=0.5,
            tools=['hover'], 
            hover_fill_color='lightblue',
            xrotation=45,  # Rotate the x-axis labels
            yformatter='%.0f',  # Ensure y-axis labels are formatted without scientific notation
        ),
        hv.opts.Overlay(
            toolbar='above',
            show_grid=True
        )
    )

# Create toggle buttons for road type, weather conditions, and pedestrian crossing facilities
road_type_toggle = pn.widgets.ToggleGroup(
    name='Road Type', 
    options=list(accidents_df['road_type'].unique()),
    button_type='success'
)

weather_toggle = pn.widgets.ToggleGroup(
    name='Weather Conditions', 
    options=list(accidents_df['weather_conditions'].unique()),
    button_type='success'
)

pedestrian_toggle = pn.widgets.ToggleGroup(
    name='Pedestrian Crossing', 
    options=list(accidents_df['pedestrian_crossing_physical_facilities'].unique()),
    button_type='success'
)

# Function to create the chart based on road type
@pn.depends(road_type_toggle.param.value)
def road_type_chart(road_types):
    filtered_data = accidents_df.filter(pl.col('road_type').is_in(road_types))
    return filtered_data.hvplot.bar(
        x='accident_severity',
        y='number_of_casualties',
        by='road_type',
        stacked=True,
        title='Casualties by Road Type and Severity',
        xlabel='Road Type',
        ylabel='Number of Casualties',
        height=500,
        width=800,
        color=color_palette
    ).opts(
        hv.opts.Bars(
            line_color='white', 
            line_width=0.5,
            tools=['hover'], 
            hover_fill_color='lightblue'
        ),
        hv.opts.Overlay(
            toolbar='above',
            show_grid=True
        )
    )

# Function to create the chart based on weather conditions
@pn.depends(weather_toggle.param.value)
def weather_chart(weather_conditions):
    filtered_data = accidents_df.filter(pl.col('weather_conditions').is_in(weather_conditions))
    return filtered_data.hvplot.bar(
        x='accident_severity',
        y='number_of_casualties',
        by='weather_conditions',
        stacked=True,
        title='Casualties by Weather Conditions and Severity',
        xlabel='Weather Conditions',
        ylabel='Number of Casualties',
        height=500,
        width=800,
        color=color_palette
    ).opts(
        hv.opts.Bars(
            line_color='white', 
            line_width=0.5,
            tools=['hover'], 
            hover_fill_color='lightblue'
        ),
        hv.opts.Overlay(
            toolbar='above',
            show_grid=True
        )
    )

# Function to create the chart based on pedestrian crossing facilities
@pn.depends(pedestrian_toggle.param.value)
def pedestrian_chart(pedestrian_crossings):
    filtered_data = accidents_df.filter(pl.col('pedestrian_crossing_physical_facilities').is_in(pedestrian_crossings))
    return filtered_data.hvplot.bar(
        x='accident_severity',
        y='number_of_casualties',
        by='pedestrian_crossing_physical_facilities',
        stacked=True,
        title='Casualties by Pedestrian Crossing Facilities and Severity',
        xlabel='Pedestrian Crossing Facilities',
        ylabel='Number of Casualties',
        height=500,
        width=800,
        color=color_palette
    ).opts(
        hv.opts.Bars(
            line_color='white', 
            line_width=0.5,
            tools=['hover'], 
            hover_fill_color='lightblue'
        ),
        hv.opts.Overlay(
            toolbar='above',
            show_grid=True
        )
    )

# Layout the toggle buttons and the dynamic charts
layout = pn.Column(
    severity_toggle, 
    filtered_chart, 
    road_type_toggle, 
    road_type_chart, 
    weather_toggle, 
    weather_chart, 
    pedestrian_toggle, 
    pedestrian_chart
)

# Display the interactive plot
layout.servable()
```
