```python
import pandas as pd
import numpy as np

# Define the display class
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    
    def __init__(self, *args):
        self.args = args

    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)

    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a)) 
                           for a in self.args)

# Create two dataframes
df1 = pd.DataFrame({
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'department': ['HR', 'IT', 'Finance', 'HR', 'IT'],
    'salary': [50000, 60000, 70000, 50000, 60000]
})

df2 = pd.DataFrame({
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Frank'],
    'hire_date': [2010, 2008, 2015, 2016, 2020],
    'bonus': [5000, 4000, 6000, 3500, 3000]
})

# Perform different types of merges
inner_join = pd.merge(df1, df2, on='employee', how='inner')
left_join = pd.merge(df1, df2, on='employee', how='left')
right_join = pd.merge(df1, df2, on='employee', how='right')
outer_join = pd.merge(df1, df2, on='employee', how='outer')

# Display the dataframes
Display = display('df1', 'df2', 'inner_join', 'left_join', 'right_join', 'outer_join')
Display
```

```python
# One-to-One Join example (similar to an inner join if all keys are unique)
one_to_one = pd.merge(df1, df2, on='employee', how='inner')
one_to_one
```

```python
# Modify df2 to create a one-to-many relationship
df2_one_to_many = pd.DataFrame({
    'employee': ['Alice', 'Alice', 'Bob', 'Charlie', 'David', 'David'],
    'hire_date': [2010, 2011, 2008, 2015, 2016, 2017],
    'bonus': [5000, 5500, 4000, 6000, 3500, 3000]
})

one_to_many = pd.merge(df1, df2_one_to_many, on='employee', how='inner')
one_to_many
```

```python
# Modify df1 to create a many-to-many relationship
df1_many_to_many = pd.DataFrame({
    'employee': ['Alice', 'Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'department': ['HR', 'Admin', 'IT', 'Finance', 'HR', 'IT'],
    'salary': [50000, 55000, 60000, 70000, 50000, 60000]
})

many_to_many = pd.merge(df1_many_to_many, df2_one_to_many, on='employee', how='inner')
many_to_many
```

```python
Display = display('df1', 'df2', 'one_to_one', 'df2_one_to_many', 'one_to_many', 'df1_many_to_many', 'many_to_many')
Display
```

### The `ON` keyword

```python
# Create two dataframes with different join keys
df3 = pd.DataFrame({
    'emp_name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'dept': ['HR', 'IT', 'Finance', 'HR', 'IT'],
    'salary': [50000, 60000, 70000, 50000, 60000]
})

df4 = pd.DataFrame({
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Frank'],
    'hire_date': [2010, 2008, 2015, 2016, 2020],
    'bonus': [5000, 4000, 6000, 3500, 3000]
})

# Perform a merge using different join keys
left_right_join = pd.merge(df3, df4, left_on='emp_name', right_on='employee', how='inner')

# Use Display class to display the original and merged DataFrames
Display = display('df3', 'df4', 'left_right_join')
Display
```

### Indexing

```python
# Assuming df1 and df2 are already defined
# Create two dataframes
df1 = pd.DataFrame({
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'department': ['HR', 'IT', 'Finance', 'HR', 'IT'],
    'salary': [50000, 60000, 70000, 50000, 60000]
})

df2 = pd.DataFrame({
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Frank'],
    'hire_date': [2010, 2008, 2015, 2016, 2020],
    'bonus': [5000, 4000, 6000, 3500, 3000]
})

# Set 'employee' as the index for both DataFrames
df1.set_index('employee', inplace=True)
df2.set_index('employee', inplace=True)

# Perform a merge using the indexes
index_merge = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')

# Use Display class to display the original and merged DataFrames
Display = display('df1', 'df2', 'index_merge')
Display
```
